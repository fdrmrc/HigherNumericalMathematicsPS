```python
import numpy as np
import matplotlib.pyplot as plt



delta = 0.025
x = y = np.arange(-3, 3, delta)
X, Y = np.meshgrid(x, y)
```

$\bullet$ Stability function $\phi: \mathbb{C} \rightarrow \mathbb{C}$ for Euler method is $\phi(z)=1+z$

```python
Z = abs(1+X+1j*Y)
plt.contour(X, Y, Z, [1], colors='k')
plt.contourf(X, Y, Z, [0,1], colors=[[1, 0.5, 0]])
plt.grid()
plt.title('Explicit Euler')
plt.show()
```

$\bullet$ Trapezoidal rule: $\phi(z)= \frac{1+\frac{z}{2}}{1- \frac{z}{2}}$

```python
Z = abs((1+0.5*(X+1j*Y))/(1-0.5*(X+1j*Y)))
plt.contour(X, Y, Z, [1], colors='k')
plt.contourf(X, Y, Z, [0,1], colors=[[1, 0.5, 0]])
plt.grid()
plt.title('Trap. Rule')
plt.show()
```

$\bullet$ Runge Kutta 2: $\phi(z)= 1 + z + \frac{z^2}{2}$

```python
Z = abs(1+(X+1j*Y)+0.5*(X+1j*Y)**2 )
plt.contour(X, Y, Z, [1], colors='k')
plt.contourf(X, Y, Z, [0,1], colors=[[1, 0.5, 0]])
plt.title('Runge Kutta 2')
plt.grid()
plt.show()
```

$\bullet$ Runge Kutta 4: $\phi(z)= 1 + z + \frac{z^2}{2} + \frac{z^3}{6} + \frac{z^4}{24}$

```python
Z = abs(1+(X+1j*Y)+0.5*(X+1j*Y)**2 + (1/6)*(X+1j*Y)**3 + (1/24)*(X+1j*Y)**4 )
plt.contour(X, Y, Z, [1], colors='k')
plt.contourf(X, Y, Z, [0,1], colors=[[1, 0.5, 0]])
plt.title('Runge Kutta 4')
plt.grid()
plt.show()
```
