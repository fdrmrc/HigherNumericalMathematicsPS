```python
import numpy as np
from numpy import sqrt
from scipy.optimize import newton_krylov
import matplotlib.pyplot as plt
```

```python
def gauss_legendre_step(rhs,tn,yn,h):
    a11 = 1/4
    a12 = 1/4 - 1/6*sqrt(3)
    a21 = 1/4 + 1/6*sqrt(3)
    a22 = 1/4
    
    b1 = 0.5
    b2 = 0.5
    
    c1 = 0.5 + 1/6*sqrt(3)
    c2 = 0.5 - 1/6*sqrt(3)
    
    def fun(p):
        x,y = p
        return np.array([x-rhs(tn+c1*h,yn+h*(a11*x+a12*y)),y-rhs(tn+c2*h,yn+h*(a21*x+a22*y))])
        #return np.concatenate([x-rhs(tn+h/3,yn+h*(5/12*x-1/12*y)),y-rhs(tn+h,yn+h*(3/4*x+1/4*y))])
    k1,k2 = newton_krylov(fun,(yn,yn),f_tol=1e-10,f_rtol=1e-6,maxiter = 100)
    return  yn + h*(b1*k1 + b2*k2)


```

```python
def rhs(t,y):
    return np.array([-3*y[0],-y[1]*(t+1)])

def uex(t):
    return np.array([np.exp(-3*t),np.exp(-0.5*t*(t+2))])

gauss_legendre_step(rhs,tn=0.1,yn=(1,1),h=0.1)
```

```python
def gauss_legendre4(rhs,t0,tf,ts,y0):
    y = np.zeros((len(y0),ts+1))
    y[:,0] = y0
    t = np.linspace(t0,tf,ts+1)
    h = (tf-t0)/ts
    for n in range(0,ts):
        y[:,n+1] = gauss_legendre_step(rhs,t[n],y[:,n],h)

    return t,y

t,y = gauss_legendre4(rhs,t0=0,tf=1,ts=10,y0=(1,1))
plt.figure()
plt.plot(t,y[1,:],'-o',t,uex(t)[1],'g')
plt.show()
```

```python
tsrange = np.arange(10,150,10)

def convergence_check(rhs,t0,tf,y0,tsrange):
    err = []
    for ts in tsrange:
        t,y = gauss_legendre4(rhs,t0,tf,ts,y0)
        err.append(np.linalg.norm(y[:,-1]-uex(tf),np.Inf))
    print(err)
    
    plt.figure(figsize=(8,8))
    plt.loglog(tsrange,err,'*',label = 'err')
    plt.loglog(tsrange,err[-1]*tsrange[-1]/tsrange,'r',label='Order 1')
    plt.loglog(tsrange,err[-1] * (tsrange[-1] / tsrange) ** 2,'b',label='Order 2')
    plt.loglog(tsrange,err[-1] * (tsrange[-1] / tsrange) ** 3,'r',label='Order 3')
    plt.loglog(tsrange,err[-1] * (tsrange[-1] / tsrange) ** 4,'g',label='Order 4')
    plt.legend()
    plt.show()

t0 = 0
tf = 1
y0 = np.array([1,1])
convergence_check(rhs,t0,tf,y0,tsrange)
```
