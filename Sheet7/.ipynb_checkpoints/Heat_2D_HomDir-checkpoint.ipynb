```python
import numpy as np
from math import cos,sin,pi
import matplotlib.pyplot as plt
from matplotlib import cm #color gradient
from mpl_toolkits.mplot3d import Axes3D  
#%matplotlib inline


# Numerical solution of the 2d heat equation on the unit square [0,1]x[0,1] with homogeneous Dirichlet boundary conditions
# with the Explicit Euler scheme in time and centered finite difference in space

#Numerical parameters
Nx = 25
Ny = 25
x = np.linspace(0,1,Nx)
y = np.linspace(0,1,Ny)
dx = x[2]-x[1]
dy = y[2]-y[1]

Tf = 0.1 #initial time
dt=np.min([dx,dy])**2/4 #stability condition for FTCS scheme
Nt = np.floor(Tf/dt).astype(int)

u=np.zeros([Nx,Ny])

def u0(x,y):
    m = 1
    n = 2
    return np.sin(m*pi*x)*np.sin(n*pi*y)#20*x*y*(x-1)*(y-1)

# def u0(x, y):
#     cond = [x<0.4 , (0.4 <= x) & (x <= 0.6), x > 0.6];
#     func = [0, 1, 0];
#     return np.piecewise(x, cond, func)

# def u0(x, y):
#     return np.piecewise(x, [x>y, x<=y], [lambda x: 1, lambda x: x**2])

# def u0(x, y):
#     cond = [x<0.4 , (0.4 <= x) & (x <= 0.6) & (0.4 <= y) & (y <= 0.6), x > 0.6, y<0.4 , y > 0.6];
#     func = [0,1,0,0,0];
#     return np.piecewise(x, cond, func)



for i in range(Nx):
    for j in range(Ny):
        u[i,j]=u0(x[i],y[j])

        
#plot initial surface
fig=plt.figure(figsize=(10,10))
x2d,y2d=np.meshgrid(x,y)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x2d, y2d, u,cmap=plt.cm.viridis)
plt.title('Initial data')

c1=dt/(dx**2);
c2=dt/(dy**2);

t=0
uNew=np.zeros([Nx,Ny])

for n in range(Nt):
    
    for i in range(1,Nx-1):
        for j in range(1,Ny-1):
            uNew[i,j]= u[i,j]+c1*(u[i,j+1]-2*u[i,j]+u[i,j-1]) + c2*(u[i+1,j]-2*u[i,j]+u[i-1,j])
            
    uNew[0,:]=np.zeros(Nx)
    uNew[:,0]=np.zeros(Nx)
    uNew[-1,:]=np.zeros(Nx)
    uNew[:,-1]=np.zeros(Nx)
    
    u=uNew.copy() #update
    t+=dt

print(Nt)
```

```python
fig=plt.figure(figsize=(14,10))
x2d,y2d=np.meshgrid(x,y)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x2d, y2d, u,cmap=plt.cm.viridis) 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u')
ax.view_init(elev=15, azim=60)
plt.title('Solution at time ' +str(np.round(Nt*dt,3)))
plt.show()
```

```python
def uex(x,y,t):
    m = 1
    n = 2
    return np.exp(-pi**2*(m**2 + n**2)*t)*np.sin(m*pi*x)*np.sin(n*pi*y)

uanal = np.zeros([Nx,Ny])
for i in range(Nx):
    for j in range(Ny):
        uanal[i,j]=uex(x[i],y[j],Tf)

#plot analytical surface
fig=plt.figure(figsize=(10,10))
x2d,y2d=np.meshgrid(x,y)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x2d, y2d, uanal)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u')
plt.show()
```

Error plot

```python
#Ntrange = np.arange(300,1900,100) # si vede che da un certo in poi  stabile
Ntrange = np.arange(50,500,50) # si vede che da un certo in poi  stabile
dtrange = Tf/Ntrange
dtrange
```

```python
err = []
k = 0
for dt in dtrange:
    Nt = Ntrange[k]
    k = k +1
    
    
#     dx = np.sqrt(4*dt)
#     dy = np.sqrt(4*dt)
#     Nx = np.ceil(1/dx).astype(int)
#     Ny = np.ceil(1/dy).astype(int)
    Nx = 25
    Ny = 25
    dx = 1/Nx
    dy = 1/Ny
    x=np.linspace(0,1,Nx)
    y=np.linspace(0,1,Ny)
    
    u=np.zeros([Nx,Ny])
    for i in range(Nx):
        for j in range(Ny):
            u[i,j]=u0(x[i],y[j])
    
    c1=dt/(dx**2);
    c2=dt/(dy**2);

    t=0
    uNew=np.zeros([Nx,Ny])

    while (abs(t-Tf)>1e-16):
        dt = np.min([dt,Tf-t])
    
        for i in range(1,Nx-1):
            for j in range(1,Ny-1):
                uNew[i,j]= u[i,j]+c1*(u[i,j+1]-2*u[i,j]+u[i,j-1]) + c2*(u[i+1,j]-2*u[i,j]+u[i-1,j])
            
        uNew[0,:]=np.zeros(Nx)
        uNew[:,0]=np.zeros(Nx)
        uNew[-1,:]=np.zeros(Nx)
        uNew[:,-1]=np.zeros(Nx)
    
        u=uNew.copy() #update
        t+=dt
    
    #print(t)
    uanal = np.zeros([Nx,Ny])
    for i in range(Nx):
        for j in range(Ny):
            uanal[i,j]=uex(x[i],y[j],Tf)

    error = np.linalg.norm(uanal - u,'fro')
    err.append(error)
```

```python
print(err)
plt.figure(figsize=(7,7))
plt.loglog(dtrange,err,'-o')#dtrange,err[-2] * dtrange[-1] / dtrange,'g')
plt.show()
```
