```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
from IPython.display import display, clear_output




def discdata(x):
    cond = [x<0.4 , (0.4 <= x) & (x <= 0.6), x > 0.6];
    func = [0, 1, 0];
    return np.piecewise(x, cond, func)



def sindata(x):
    return 6*np.sin(pi*x)

def uex(x,t):
    return 6*np.sin(pi*x)*np.exp(-pi**2*t)


initialdata = sindata
x = np.linspace(0,1,10+1)
u0 = initialdata(x)
plt.figure()
plt.plot(x,u0,'g')
plt.title('initial data')
plt.show()

```

```python
# Building the mass matrix and the stiffness matrix

def stiffassembly(M):
    x = np.linspace(0,1,M+1)
    diag = np.zeros(M-1) #x_1,...,x_M-1 (M-1)
    subd = np.zeros(M-2) 
    supr = np.zeros(M-2)
    h = np.diff(x)
    for i in range(1,M):
        diag[i-1] = 1/h[i-1] +1/h[i]

    for k in range(1,M-1):
        supr[k-1] = -1/h[k]
        subd[k-1] = -1/h[k]

    A = np.diag(subd,-1) + np.diag(diag,0) + np.diag(supr,+1)
    return A


def massmatrix(N):
    x = np.linspace(0,1,N+1)
    diag = np.zeros(N-1) #x_1,...,x_M-1 (M-1)
    subd = np.zeros(N-2) 
    supr = np.zeros(N-2)
    h = np.diff(x)
    for i in range(1,N):
        diag[i-1] = (h[i-1] + h[i])/3

    for k in range(1,N-1):
        supr[k-1] = h[k]/6
        subd[k-1] = h[k-1]/6

    M = np.diag(subd,-1) + np.diag(diag,0) + np.diag(supr,+1)
    return M


def fem_heat(tf,ts,initialdata):
    #tf: final time
    #ts: timesteps
    #initialdata
    dt = tf/ts
    h = np.sqrt(10*dt)
    N = (1/h).astype(int)
    print(N)
    x = np.linspace(0,1,N+1)
    #h = 1/N
    #dt = 0.1*h**2
    #ts = np.floor(tf/dt).astype(int)

    M = massmatrix(N) 
    A = stiffassembly(N)
    u0 = initialdata(x[1:-1])
    u = np.zeros([N-1,ts+1])
    u[:,0] = u0

    #u = u0
    #un = u.copy()
    for i in range(0,ts):
        u[:,i+1] = np.linalg.solve(M,np.matmul(M-dt*A, u[:,i]) )
        #u[:,i+1] = np.linalg.solve(M +dt*A,np.matmul(M, u[:,i]) ) #        un = u.copy()
    return x, u
```

```python
tf = 0.1

x, u = fem_heat(tf,100,initialdata)
plt.figure()
plt.plot(x[1:-1],u[:,-1],'-o',label='numerical')
plt.plot(x[1:-1],uex(x[1:-1],tf),label='analytical')
plt.xlabel('x')
plt.title('$u(T_f,x)$')
plt.legend()
plt.show()
plt.show()
```

Convergence

```python
def convergence_plot(tsrange,uex):
    err = []
    for ts in tsrange:
        tf = 0.2
        x, u = fem_heat(tf,ts,sindata)
        err.append(np.linalg.norm(u[:,-1] - uex(x[1:-1],tf),np.Inf))
    plt.figure()
    plt.loglog(tsrange,err,'-o',label='err')
    plt.loglog(tsrange,err[-1]*tsrange[-1]/tsrange,'r',label='Order 1')
    plt.loglog(tsrange,err[-1]*(tsrange[-1]/tsrange)**2,'g',label='Order 2')
    plt.legend()
    plt.show()
    
Backtsrange = np.arange(100,1000,100)#
convergence_plot(Backtsrange,uex)
```
